### Handling Missing Values

```python
# total missing values of each column
df.isnull().sum()
```

```python
#total missing values
(df.isnull().sum()).sum()
```

```python
# 1.Gender Column
df["Gender"].unique()
```

```python
df["Gender"].mode()
```

```python
df["Gender"]=df["Gender"].replace(np.nan,("Male"))
```

```python
# 2.Married column
df["Married"].unique()
```

```python
df["Married"].mode()
```

```python
df["Married"]=df["Married"].replace(np.nan,("Yes"))
```

```python
# Dependents colunn
print(df["Dependents"].unique())
print()
print(df["Dependents"].mode())
```

```python
df["Dependents"]=df["Dependents"].replace(np.nan,("0"))
```

```python
#replacing value
df["Dependents"]=df["Dependents"].replace({"3+":"3"}) #df["Dependents"].replace("3+",3,inplace=True)
```

```python
# self employed column
print(df["Self_Employed"].unique())
print()
print(df["Self_Employed"].mode())
```

```python
df["Self_Employed"]=df["Self_Employed"].replace(np.nan,("No"))
```

```python
# Loan Amount Column
print(df["LoanAmount"].unique())
print()
print(df["LoanAmount"].median())
```

```python
df["LoanAmount"]=df["LoanAmount"].replace(np.nan,("129"))
```

```python
# Loan_Amout_Term
print(df["Loan_Amount_Term"].unique())
print()
print(df["Loan_Amount_Term"].mode())
```

```python
print(df["Loan_Amount_Term"].unique())
```

```python
df["Loan_Amount_Term"]=df["Loan_Amount_Term"].replace(np.nan,("360"))
```

```python
#Credit_History
print(df["Credit_History"].unique())
print()
print(df["Credit_History"].mode())
```

```python
df["Credit_History"]=df["Credit_History"].replace(np.nan,(1))
```

### Changing Dtype

```python
df.info()
```

```python
# changing dtype
df["Dependents"]=df["Dependents"].astype(int)
df["CoapplicantIncome"]=df["CoapplicantIncome"].astype(int)
df["LoanAmount"]=df["LoanAmount"].astype(int)
df["Loan_Amount_Term"]=df["Loan_Amount_Term"].astype(int)
df["Credit_History"]=df["Credit_History"].astype(int)
```

```python
df.info()
```

### Column Transformation

```python
# replacing column name
df.rename(columns={'Married':'Marital_Status'},inplace=True)
df.rename(columns={'Loan_Amount_Term':'Loan_Amount_Term(years)'},inplace=True)
```

```python
df["Marital_Status"]=df["Marital_Status"].replace({"Yes":"Married"})
df["Marital_Status"]=df["Marital_Status"].replace({"No":"Unmarried"})
```

```python
df['Loan_Amount_Term(years)']=df['Loan_Amount_Term(years)'].replace({12:1})
df['Loan_Amount_Term(years)']=df['Loan_Amount_Term(years)'].replace({36:3})
df['Loan_Amount_Term(years)']=df['Loan_Amount_Term(years)'].replace({60:5})
df['Loan_Amount_Term(years)']=df['Loan_Amount_Term(years)'].replace({87:7})
df['Loan_Amount_Term(years)']=df['Loan_Amount_Term(years)'].replace({120:10})
df['Loan_Amount_Term(years)']=df['Loan_Amount_Term(years)'].replace({180:15})
df['Loan_Amount_Term(years)']=df['Loan_Amount_Term(years)'].replace({240:20})
df['Loan_Amount_Term(years)']=df['Loan_Amount_Term(years)'].replace({300:25})
df['Loan_Amount_Term(years)']=df['Loan_Amount_Term(years)'].replace({360:30})
df['Loan_Amount_Term(years)']=df['Loan_Amount_Term(years)'].replace({480:40})
```

## TRAINING AND TESTING

```python
from sklearn.model_selection import train_test_split
```

```python
x=df_2.iloc[:,:-1]
y=df_2["Loan_Status"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3,random_state=42)
```

## MODEL BUILDING

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.pred=lr.predict([[0,1,0,0,1,0.238250,-0.933094,0.123329,0.262394,0,1]])
result = lr.pred 
if result == 1:
    print("Approved")
else:
    print("Not_Approved")
```
